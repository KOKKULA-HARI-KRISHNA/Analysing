# 1 - Explore the dataset and fix the data structure and presentation.*

```python
# IMPORTING LIBRARIES PANDAS AS PD AND NUMPY AS NP
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
import pandas as pd
import numpy as np
data = pd.read_csv('DataScienceCleanStatement.csv') # IMPORTING CSV FILE
data.dtypes # CHECKING DATA TYPES
```

```python
import pyttsx3
eng = pyttsx3.init()
eng.say('Successfully installed data')
eng.runAndWait()
```

```python
data['Category'] = data['Category'].replace(r'^\s+$', np.nan, regex=True)
```

```python
data['Category']
```

```python
data.shape # CHECKING THE SHAPE
```

```python
# seperating 'object' datatype columns and 'float' datatype columns
str_data = data.select_dtypes(include = object)
num_data = data.select_dtypes(include = np.number)
str_data.columns
```

```python
# removing white spaces before and after the column records
for i in str_data.columns:
    str_data[i] = str_data[i].str.strip()
```

```python
# concating the str_data ad num_data
data_1 = pd.concat([str_data,num_data], axis = 1)
```

```python
# COLUMN DATE HAS DATA TYPE 'OBJECT'.., SO WE SHOULD CONVERT INT0 DATE TIME DATA TYPE
data_1['Date'] = pd.to_datetime(data['Date'])
```

```python
# selecting column 'country' by deleting null records and by droping duplicates of 'city' column and keeping index as 'city'
g = data_1.dropna(subset = ['Country']).drop_duplicates('City').set_index('City')['Country']
```

```python
# filling the null values of country by mapping with city olumn
data_1['Country'] = data_1['Country'].fillna(data_1['City'].map(g))
```

```python
data_1['Locality'] = data_1['Locality'].fillna(data_1['TransactionUserName'].map(k))
```

```python
data_1.isnull().sum()
```

```python
l = data_1.dropna(subset = ['State']).drop_duplicates('TransactionUserName').set_index('TransactionUserName')['State']
data_1['State'] = data_1['State'].fillna(data_1['TransactionUserName'].map(l))
```

```python
data_1.isnull().sum()
```

```python
k = data_1.dropna(subset = ['City']).drop_duplicates('State').set_index('State')['City']
data_1['City'] = data_1['City'].fillna(data_1['State'].map(k))
```

2) Compare my monthly travel over the past three month

```python
data_1.columns
```

```python
data_1[(data_1['Date'].dt.year == 2020) & (data_1['Date'].dt.month >= 10)].head(1)
```

```python
grp = data_1.groupby(data_1[(data_1['Date'].dt.year == 2020) & (data_1['Date'].dt.month >= 10)]['Date'].dt.month)
A = grp.agg({'Amount':'sum'})
data_A = A.reset_index()
data_A = data_A.rename(columns = {'Date':'MONTH'})
data_A
```

```python
data_A['Amount'] = data_A['Amount']* -1
```

```python
import seaborn as sns
sns.catplot(data=data_A, kind="bar", x="MONTH", y="Amount")
```

```python
g = sns.relplot(x = 'MONTH', y = 'Amount', kind = 'line', data = data_A)
```
