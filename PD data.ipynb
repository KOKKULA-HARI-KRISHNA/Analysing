# 2) inter quartile range

```python
import numpy as np
```

```python
def outliers_data(column):
    sort_data = sorted(data[column])
    quantile1, quantile3 = np.percentile(sort_data,[25,75])
    inter_quantile_range = quantile3-quantile1
    lower_bound_value = quantile1 - (1.5 * inter_quantile_range)
    upper_bound_value = quantile3 + (1.5 * inter_quantile_range)
    print('lower_bound --------->',lower_bound_value)
    print('upper_bound --------->',upper_bound_value)
    print()
    print('data below lower bound')
    print(data[data[column] < lower_bound_value][column])
    print('data above upper bound')
    print(data[data[column] > upper_bound_value][column])
```

```python
pd.options.display.max_rows = None
pd.options.display.max_columns = None
np.set_printoptions(suppress=True)

```

```python
for i in data.columns:
    print('outliers data of column', i)
    print()
    outliers_data(i)
    print()
```

# 3) using z-score

```python
outliers = []
def outliers_z_score(i):
    threshold1 = 3
    threshold2 = -3
    mean = np.mean(data[i])
    std = np.std(data[i])
    for i in data[i]:
        z_score = (i - mean)/ std
        if np.abs(z_score) > threshold1 or np.abs(z_score) < threshold2:
            outliers.append(i)
    return outliers
```

```python
data.columns
```

```python
for i in data.columns:
    print('column name --> ',i)
    print('outliers detected:- ')
    outliers_z_score(i)
    print(outliers)
    del outliers[:]
    print()
```
